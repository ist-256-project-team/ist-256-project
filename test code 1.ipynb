```python
#first person (english to chinese)+second person
#test
#one loop
language_b = "zh-CN"
language_a = "en"
import speech_recognition as sr
import time


#first person
while True:
    try:
        r = sr.Recognizer()
        with sr.Microphone() as source:
            print("Say something!")
            audio = r.listen(source)
            text_from_langinput_a = r.recognize_google(audio,language=language_a)
            print("You said: " + text_from_langinput_a)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        text_from_langinput_a='Google Speech Recognition could not understand audio'
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))


    #translation API 


    from google.cloud import translate

    translate_client = translate.Client()

    
    if text_from_langinput_a=='goodbye':
        break
    if text_from_langinput_a=='Google Speech Recognition could not understand audio':
        continue
    target = language_b
    translation = translate_client.translate(
        text_from_langinput_a,
        target_language=target)

    
    print(u'Translation: {}'.format(translation['translatedText'])) 
    translated_text=translation['translatedText']
    
    from gtts import gTTS
    tts=gTTS(translated_text,lang=language_b)
    tts.save('new.mp3')
    from pygame import mixer
    mixer.init()
    mixer.music.load('new.mp3')
    mixer.music.play()
    
    time.sleep(4)
 #second person   
    try:
        r = sr.Recognizer()
        with sr.Microphone() as source:
            print("Say something!")
            audio = r.listen(source)
            text_from_langinput_b = r.recognize_google(audio,language=language_b)
            print("You said: " + text_from_langinput_b)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
        text_from_langinput_b='Google Speech Recognition could not understand audio'
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))


    #translation API 


    from google.cloud import translate

    translate_client = translate.Client()

    

    target = language_a
    translation = translate_client.translate(
        text_from_langinput_b,
        target_language=target)

    
    print(u'Translation: {}'.format(translation['translatedText'])) 
    translated_text=translation['translatedText']
    if translated_text=='Goodbye':
        break
    if translated_text=='Google Speech Recognition could not understand audio':
        continue 
    
    from gtts import gTTS
    tts=gTTS(translated_text,lang=language_a)
    tts.save('new.mp3')
    from pygame import mixer
    mixer.init()
    mixer.music.load('new.mp3')
    mixer.music.play()
    time.sleep(4)
```

```python
#part 1
import pandas as pd
data = pd.read_html('https://cloud.google.com/translate/docs/languages')
language_df=data[0]

language_dict = language_df.to_dict(orient='records')
#get a list of full language name, the list is called ful_lag
ful_lag=[]
for language in language_dict:
    ful_lag.append(language['Language'])
ful_lag[13]='Chinese_S'
ful_lag[14]='Chinese_T'
#ful_lag
#get a list of abbreviated language name, the list is called abr_lag
abr_lag=[]
for language in language_dict:
    abr_lag.append(language['ISO-639-1 Code'])
#abr_lag
#make a dictionary with "full language name" as the key and "abbreviated language name" as the value
lag_dict={}
for n in range(len(abr_lag)):
    lag_dict[ful_lag[n]]=abr_lag[n]
lag_dict['Chinese_S']='zh-CN'
lag_dict['Chinese_T']='zh-TW'

#code body
while True:   
    language_one=input('Before you begin your phone call, enter your language: ')
    if language_one=='stop':
        break
    if language_one in lag_dict:  
        ISO_Code_1=lag_dict[language_one]
        language_a = ISO_Code_1
        print(ISO_Code_1)
        break
    else:
        print('Invalid Entry, please try again.')
while True: 
    if language_one != 'stop':
        language_two=input('Before you begin your phone call, enter the language of the recipient: ')
        if language_two=='stop':
            break
        if language_two in lag_dict:  
            ISO_Code=lag_dict[language_two]
            language_b = ISO_Code
            print(ISO_Code)
            break
        else:
            print('Invalid Entry, please try again.')
```
