```python
try:
    import urllib.request as urllib2
except ImportError:
    import urllib2

def text_to_speech(text,ISO):
    from gtts import gTTS
    tts=gTTS(text,lang=ISO)
    tts.save('new.mp3')
    from pygame import mixer
    mixer.init()
    mixer.music.load('new.mp3')
    mixer.music.play()

def translation(text,ISO):
    from google.cloud import translate

    translate_client = translate.Client()


    translation = translate_client.translate(
        text,
        target_language=ISO)
    translated_text=translation['translatedText']
    return translated_text


def get_ISO_code(your_language):
      
    import pandas as pd
    data = pd.read_html('https://cloud.google.com/translate/docs/languages')
    language_df=data[0]

    language_dict = language_df.to_dict(orient='records')
    #get a list of full language name, the list is called ful_lag
    ful_lag=[]
    for language in language_dict:
        ful_lag.append(language['Language'])
    ful_lag[13]='Chinese_S'
    ful_lag[14]='Chinese_T'
    #ful_lag
    #get a list of abbreviated language name, the list is called abr_lag
    abr_lag=[]
    for language in language_dict:
        abr_lag.append(language['ISO-639-1 Code'])
    #abr_lag
    #make a dictionary with "full language name" as the key and "abbreviated language name" as the value
    lag_dict={}
    for n in range(len(abr_lag)):
        lag_dict[ful_lag[n]]=abr_lag[n]
    lag_dict['Chinese_S']='zh-CN'
    lag_dict['Chinese_T']='zh-TW'

    ISO_Code=lag_dict[your_language]
    return(ISO_Code)

try:
    
    while True: 

        language_a=input('Before you begin your phone call, enter your language : ')
        language_a=language_a.title()
        
        try:
            
            ISO_Code_a=get_ISO_code(language_a)
            print(ISO_Code_a)
            break
        except KeyError:
            print('Invalid Entry, please try again.')
    
    while True: 
        
        language_b=input("Before you begin your phone call, enter the recipient's language : ")
        language_b=language_b.title()
        
        
        try:
    
            
            ISO_Code_b=get_ISO_code(language_b)
            print(ISO_Code_b)
            break
        except KeyError: 
            print('Invalid Entry, please try again.')
    #part2           


    import speech_recognition as sr
    import time

    #first person language(a==>b)
    #take in language a
    while True:
        
        try:
            
            r = sr.Recognizer()
            with sr.Microphone() as source:
                print("Say something in %s: "% language_a)
                audio = r.listen(source)
                text_from_langinput_a = r.recognize_google(audio,language=ISO_Code_a)
                print("You said: " + text_from_langinput_a)
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
            text_from_langinput_a='Google Speech Recognition could not understand audio'
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))

    #check if it is 'goodbye'

        if text_from_langinput_a=='Google Speech Recognition could not understand audio':
            continue

        if translation(text_from_langinput_a,'en').lower()=='goodbye':
            print('Goodbye')
            print('Called ended...')
            break

        #translate to language b 

        translated_text_b=translation(text_from_langinput_a,ISO_Code_b)
        print('Translation: %s'%translated_text_b)

        # read in language b
        text_to_speech(translated_text_b,ISO_Code_b)       
        time.sleep(5)
        #second person  language (b==>a)
    #take in language b
        try:
            r = sr.Recognizer()
            with sr.Microphone() as source:
                print("Say something in %s: "%language_b)
                audio = r.listen(source)
                text_from_langinput_b = r.recognize_google(audio,language=ISO_Code_b)
                print("You said: " + text_from_langinput_b)
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
            text_from_langinput_b='Google Speech Recognition could not understand audio'
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))

        if text_from_langinput_b=='Google Speech Recognition could not understand audio':
            continue
    #check if it is 'goodbye'    

        if translation(text_from_langinput_b,'en').lower()=='goodbye':
            print('Goodbye')
            print('Called ended...')
            break

    #translate to language a
        translated_text_a=translation(text_from_langinput_b,ISO_Code_a)
        print('Translation: %s'%translated_text_a)

     #read in language a   
        text_to_speech(translated_text_a,ISO_Code_a)
        time.sleep(5)
        try:
            import urllib.request as urllib2
        except ImportError:
            import urllib2
except urllib2.URLError as e:
    print("ERROR: Response from URL was not ok.")
    print("DETAILS:", e)
        
```
